```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('UNPD_WCU2019_Country_Data_Survey-Based.xlsx',
                     sheetname="By methods", header=4, usecols=range(22)).reset_index()
```

```python
data.rename(columns={"level_0":"Country or area", "level_1":"ISO code",
                     "level_2":"Survey start year", "level_3":"Survey end year",
                     "level_4":"Age group"}, inplace=True)
```

```python
data.set_index(['Country or area', 'ISO code'], inplace=True)
```

```python
data.replace("..", np.nan, inplace=True)
```

```python
most_recent_idx_end = data.groupby(['Country or area'])['Survey end year'].transform(max) == data['Survey end year']
most_recent_data = data[most_recent_idx_end]
```

```python
most_recent_idx_start = most_recent_data.groupby(['Country or area'])['Survey start year'].transform(max) == most_recent_data['Survey start year']
most_recent_data_start = most_recent_data[most_recent_idx_start]
```

```python
most_recent_idx_source = most_recent_data_start.groupby(['Country or area'])['Source'].transform(max) == most_recent_data_start['Source']
most_recent_data_source = most_recent_data_start[most_recent_idx_source]
```

```python
most_recent_data_source.sort_values("Any modern method",
                             ascending=False)[['Survey start year', 'Survey end year',
                                               'Age group', 'Any modern method']]
```
